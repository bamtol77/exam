```python
#전체 데이터 컬럼명에서 범주형 데이터 만을 추출
total_list = df1.columns.tolist() #전체 컬럼명 리스트
numeric_list = df1.describe().columns.tolist() #연속형데이터
object_list = list(set(total_list)-set(numeric_list)) #연속형 list에서 범주형 list를 제외한 나머지
```

```python
df1[object_list].describe()
```
