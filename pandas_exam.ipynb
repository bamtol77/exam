# 데이터 필터
- 특정조건을 걸어서 추출

```python
# 연령이 40대만 추출하고 싶은 경우
df1['연령대'].value_counts() # 연령대별로 몇명이 있는지 확인
```

```python
df1['연령대'] == '40대' # 40대인 사람들만 추출하는 조건
```

```python
cond1 = (df1['연령대']== '40대')
```

```python
df1.loc[cond1] # 조건을 걸어서 추출 #조건이 맞는 행을 모두 출력
```

```python
df1.loc[cond1].shape
```

```python
#40대를 제외한 나머지 연령대의 사람들을 추출하고 싶은 경우
cond1 = (df1['연령대']!= '40대')
df1.loc[cond1] # 조건을 걸어서 추출 #조건이 맞는 행을 모두 출력
```

```python
#40대를 제외한 나머지 연령대의 사람들을 추출하고 싶은 경우
cond1 = (df1['연령대']== '40대')
df1.loc[~cond1] # '~'을 사용해서 반대대는 결과를 얻을 수 있음 조건을 걸어서 추출 #조건이 맞는 행을 모두 출력
```

```python
# 40대 이면서 남성인 사람들을 추출하고 싶은 경우
df1['성별'].value_counts() # 성별별로 몇명이 있는지 확인
```

```python
cond1 = (df1['연령대'] == '40대') # 40대인 사람들만 추출하는 조건
cond2 = (df1['성별'] == '남성') # 남성인 사람들만 추출하는 조건
df1.loc[cond1 & cond2] #or 조건은 "|"을 사용한다 # 조건을 걸어서 추출 #조건이 맞는 행을 모두 출력

```

# 연습
 - 30대 여성 추출 -> 구매금액 높은 순서로 정렬
 - 파일로 저장

```python
cond1 = (df1['연령대'] == '30대이하')
cond2 = (df1['성별'] == '여')

df1[cond1 & cond2]
```

```python
#구매금액이 높은 순서로 정렬하고 싶은 경우
df3 = df1[cond1 & cond2].sort_values(by='구매금액', ascending=False) # 구매금액을 높은순으로 정렬
df3.shape
```

```python
df3.to_csv('result.csv', encoding='cp949')  #한글이 깨지지 않게 하기 위해서는 encoding='utf-8-sig'를 사용한다.
```
