```python
df1.iloc[ :, 8:-2].T # Transpose: 행과 열을 바꾼다.
#문제는 column이 너무 많아서 가독성이 떨어진다.
```

* .T를 했을 때 문제점을 해결하기 위해서 Stack 함수를 사용
* 데이터를 Column형태로 쌓아준다

```python
df2 = df1.iloc[ :, 8:-2]
```

```python
df2.stack() # stack() 함수는 DataFrame을 Series로 변환하며, 행과 열을 재구성한다. // Series형태로 변환
```

```python
pd.DataFrame(df2.stack()) # stack을 하면 column이 사라지고 # index가 생긴다. 
# Column이 사라지면서 index가 생기고, 행과 열이 바뀐다.
```

```python
pd.DataFrame(df2.stack()).reset_index() # reset_index()를 사용하여 인덱스를 초기화한다.
# index에 있는 값이 column으로 바뀐다.
```

```python
pd.DataFrame(df2.stack()).reset_index().iloc[1:] # iloc[1:]을 사용하여 첫 번째 행을 제외한다. // 첫번째 행은 index가 된다.
```

# 두 데이터 프레임 붙치는 방법 (4가지)
* Table Join
    - right /left ;  
    - inner ; 중복되어 있는 것들만 합침 (교집합 이라고 생각)
    - outer ; 중복되어 있지 않는 것들 합침


```python
df5.head(2)
```

```python
#df5 와 df6을 합친다.
# 제품명을 기준으로 합친다.
# 날짜 데이터를 Key로 사용하기 위해서 df5기준으로 합친다.
pd.merge(df6,df5, how='right', on='제품명') # right join을 사용하여 df5를 기준으로 합친다.
```

```python
df7 = pd.merge(df6,df5, how='right', on='제품명') 
df7.to_excel('total_data.xlsx')#저장
```

```python
pd.to_datetime(df7['날짜'])
```

```python
df7['날짜(전처리)'] = pd.to_datetime(df7['날짜'])
df7 = df7.drop(columns=['공급원'])
df7.head(2)
```

```python
df7['공급월'] = df7['날짜(전처리)'].dt.month
df7.head(2)
```

```python
pd.pivot_table(data = df7, index = '공급월', values='재고량', aggfunc='sum')
```
